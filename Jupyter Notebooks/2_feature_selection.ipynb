### Feature selection

For modeling we will be only using those features which are available at the initial stage of the loan.


```python
features = ['originalLoanAmount','originalLoanTerm','paymentToIncomePercentage','originalInterestRatePercentage' #metrics
          ,'vehicleManufacturerName','vehicleNewUsedCodeM','vehicleModelYear','vehicleValueAmount','vehicleValueSourceCodeM' #vehicle
            ,'coObligorIndicator','obligorCreditScore','obligorGeographicLocation','underwritingIndicator','obligorCreditScoreType' #obligor
            ,'obligorEmploymentVerificationCodeM','vehicleTypeCodeM','obligorIncomeVerificationLevelCodeM']
```

```python
#simplifying column names
cols={x+'LocRandom':x for x in features}
data.rename(columns=cols,inplace=True)
```

```python
include=['reportingPeriodBeginningDateREvent','reportingPeriodEndingDateREvent','ltv']
```

```python
data=data[features+include+['binaryTargetNumeric']]
```

```python
data.head()
```

### Handle na

```python
data.isnull().sum()
```

```python
credit_mean=data['obligorCreditScore'].mean()
data['obligorCreditScore']=data['obligorCreditScore'].fillna(credit_mean)
```
